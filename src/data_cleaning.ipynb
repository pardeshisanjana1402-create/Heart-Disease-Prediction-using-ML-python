```python
import pandas as pd

# Load files
labels_path = r"C:\Users\Sanjana\Downloads\labels.csv"
values_path = r"C:\Users\Sanjana\Downloads\values.csv"

labels = pd.read_csv(labels_path)
values = pd.read_csv(values_path)

print("Labels Shape:", labels.shape)
print("Values Shape:", values.shape)
```

```python
#labels------

# Remove duplicate rows
labels_clean = labels.drop_duplicates()

# Strip column names
labels_clean.columns = labels_clean.columns.str.strip()

# Replace blank/whitespace with NaN
labels_clean.replace(r'^\s*$', pd.NA, regex=True, inplace=True)
```

```python
#values----------
values_clean = values.drop_duplicates()
values_clean.columns = values_clean.columns.str.strip()
values_clean.replace(r'^\s*$', pd.NA, regex=True, inplace=True)

print("Basic cleaning complete!")
```

```python
# Change the key if needed
common_key = "patient_id"

merged_df = pd.merge(labels_clean, values_clean, on=common_key, how="inner")

print("Merged Shape:", merged_df.shape)
merged_df.head()
```

```python
# Handle missing values
merged_df = merged_df.fillna(merged_df.median(numeric_only=True))

# For categorical columns
cat_cols = merged_df.select_dtypes(include=['object']).columns
merged_df[cat_cols] = merged_df[cat_cols].fillna("Unknown")

# Remove duplicates again (safety)
merged_df = merged_df.drop_duplicates()

# Reset index
merged_df.reset_index(drop=True, inplace=True)

print("Final cleaning done!")
merged_df.head()
```

```python
save_path = r"C:\Users\Sanjana\Downloads\final_cleaned.csv"
merged_df.to_csv(save_path, index=False)

print("File saved at:", save_path)
```
