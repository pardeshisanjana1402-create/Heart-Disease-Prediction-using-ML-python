### **Visualization**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Sanjana\Downloads\final_cleaned.csv")

target = None
for c in df.columns:
    if df[c].nunique() == 2:
        target = c
        break

numeric_cols = df.select_dtypes(include=[np.number]).columns
age_col = "age" if "age" in df.columns else numeric_cols[0]
chol_col = "chol" if "chol" in df.columns else numeric_cols[1]
thalach_col = "thalach" if "thalach" in df.columns else numeric_cols[2]

plt.figure(figsize=(6,4))
df[target].value_counts().plot(kind='bar')
plt.title(f"Distribution of {target}")
plt.xlabel(target)
plt.ylabel("Count")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
plt.hist(df[age_col], bins=20)
plt.title(f"Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

if "sex" in df.columns:
    plt.figure(figsize=(6,4))
    df.groupby("sex")[target].mean().plot(kind="bar")
    plt.title("Gender vs Heart Disease Rate")
    plt.xlabel("Gender (0=F, 1=M)")
    plt.ylabel("Heart Disease Rate")
    plt.tight_layout()
    plt.show()

plt.figure(figsize=(4,6))
plt.boxplot(df[chol_col].dropna())
plt.title("Cholesterol Boxplot")
plt.ylabel(chol_col)
plt.tight_layout()
plt.show()

corr = df[numeric_cols].corr()
plt.figure(figsize=(8,6))
im = plt.imshow(corr)
plt.colorbar(im)
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df[age_col], df[thalach_col])
plt.xlabel(age_col)
plt.ylabel(thalach_col)
plt.title(f"{age_col} vs {thalach_col}")
plt.tight_layout()
plt.show()
```


## 1. Target Variable Distribution (Bar Chart)
**What it shows:**  
This bar chart shows how many patients have heart disease (1) and how many do not (0).

**Why used:**  
- Helps check dataset balance  
- Shows frequency of heart disease  
- Important for ML model understanding
